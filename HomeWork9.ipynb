```python
# Завантаження необхідних бібліотек
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd

# Створимо тестові дані для демонстрації
data_2d = pd.DataFrame({
    'Feature 1': np.random.rand(100),
    'Feature 2': np.random.rand(100)
})

# Функція для визначення оптимальної кількості кластерів
def elbow_method(data):
    distortions = []
    K = range(1, 11)
    for k in K:
        kmeanModel = KMeans(n_clusters=k)
        kmeanModel.fit(data)
        distortions.append(kmeanModel.inertia_)

    plt.figure(figsize=(8, 4))
    plt.plot(K, distortions, 'bx-')
    plt.xlabel('Number of clusters')
    plt.ylabel('Distortion')
    plt.title('Elbow Method showing the optimal k')
    plt.show()

# Виклик функції для тестових даних
elbow_method(data_2d)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np

# Sample data since the original data is not provided
data_mnist = pd.DataFrame(np.random.rand(100, 784))

# Функція для визначення оптимальної кількості кластерів
def elbow_method(data):
    distortions = []
    K = range(1, 11)
    for k in K:
        kmeanModel = KMeans(n_clusters=k)
        kmeanModel.fit(data)
        distortions.append(kmeanModel.inertia_)

# Scale the MNIST data
scaler = StandardScaler()
mnist_scaled = scaler.fit_transform(data_mnist)  # Scale and store the result

# Зменшення розмірності за допомогою PCA
pca = PCA(n_components=2)
mnist_pca = pca.fit_transform(mnist_scaled)  # Now mnist_scaled is defined

# Виклик функції для даних MNIST
elbow_method(mnist_pca) # Now mnist_pca is defined in this scope

# Виконання кластеризації для MNIST даних
kmeans_mnist = KMeans(n_clusters=10)
mnist_clusters = kmeans_mnist.fit_predict(mnist_pca)

# Візуалізація результатів кластеризації для MNIST даних
plt.figure(figsize=(8, 6))
plt.scatter(mnist_pca[:, 0], mnist_pca[:, 1], c=mnist_clusters, cmap='viridis') # mnist_pca is now defined
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering (MNIST Data Reduced with PCA)')
plt.show()
```
